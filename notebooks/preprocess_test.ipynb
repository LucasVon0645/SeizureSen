## Joining Pre-Processed Data

```python
data_directory_2 = project_root / "data" / "preprocessed" / "Dog_1_2_5s_slices"

data_preprocessed_2 = load_preprocessed_data(data_directory_2, file_name)

X_2 = data_preprocessed_2["X"]
y_2 = data_preprocessed_2["y"]

print("Number of slices: ", len(X_2))

print(X_2)
```

```python
X_joined = np.concatenate((X_2, X))
y_joined = np.concatenate((y_2, y))

print("Number of slices: ", len(X_joined))
print("Number of labels: ", len(y_joined))
```
